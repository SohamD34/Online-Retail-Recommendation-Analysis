```python
# Classifying on the basis of months

L = list(df["InvoiceDate"])
month_dict = {"12-2010":1,"01-2011":2,"02-2011":3,"03-2011":4,"04-2011":5,"05-2011":6,"06-2011":7,"07-2011":8,"08-2011":9,"09-2011":10,"10-2011":11,"11-2011":12,"12-2011":13}

for i in range(len(L)):
    L[i] = (L[i])[3:10]

month_wise_class = []
for i in L:
    month_wise_class.append(month_dict[i])

df['month_wise_class'] = month_wise_class
df
```

```python
# BASED ON HOURS OF THE DAY

L = list(df["InvoiceDate"])

for i in range(len(L)):
    L[i] = int((L[i])[11:13])

bins = []

for i in L:
    if 6<=i and i<=8:
        bins.append(1)
    elif 9<=i and i<=11:
        bins.append(2)
    elif 12<=i and i<=14:
        bins.append(3)
    elif 15<=i and i<=17:
        bins.append(4)
    elif 18<=i and i<=20:
        bins.append(5)

df["Time-bins"] = bins
df
```

```python
# BASED ON PRICE

L = list(df["UnitPrice"])
drops = []

for i in range(len(L)):
    if L[i] < 0:
        drops.append(i)

df = df.drop(drops)

m = min(df["UnitPrice"])
M = max(df["UnitPrice"])

df
```

```python
money_bins = []
A = list(df["UnitPrice"])

for i in A:
    if 0<=i and i<=5000:
        money_bins.append(1)
    elif 5001<=i and i<=10000:
        money_bins.append(2)
    elif 10001<=i and i<=15000:
        money_bins.append(3)
    elif 15001<=i and i<=20000:
        money_bins.append(4)
    elif 20001<=i and i<=25000:
        money_bins.append(5)
    elif 25001<=i and i<=30000:
        money_bins.append(6)
    elif 30001<=i and i<=35000:
        money_bins.append(7)
    elif 35001<=i and i<=40000:
        money_bins.append(8)

df["Money_bins"] = money_bins
df
```
